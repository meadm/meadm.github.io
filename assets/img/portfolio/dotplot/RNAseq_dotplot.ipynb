# Making a dotplot for the RNA-seq case study on the Prairie Helix Consulting Website

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate data

```python
np.random.seed(42) 
```

```python
# Create 3 clusters
cluster1 = np.random.normal(loc=[0,0], scale=1.0, size=(20,2))
cluster2 = np.random.normal(loc=[5,5], scale=1.0, size=(20,2))
cluster3 = np.random.normal(loc=[0,5], scale=1.0, size=(20,2))
```

```python
# Combine all clusters
data = np.vstack([cluster1, cluster2, cluster3])
```

## Assign colors

```python
colors = ['#66c2a5']*20 + ['#ef8a62']*20 + ['#8da0cb']*20  # green, orange, blue
```

## Make Plot

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(data[:,0], data[:,1], c=colors, s=80, edgecolor='k')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
plt.close(fig)
```
